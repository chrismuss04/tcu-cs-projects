```python
import math
import pandas as pd


def find_max_subarray(input, low, high):
    if (low == high):
        return (low, high, input[low])
    else:
        mid = (low + high) // 2
        #Find the max subarray for the lower 1/2, upper 1/2, and the cross between the two
        ll, lh, ls = find_max_subarray(input, low, mid)
        rl, rh, rs = find_max_subarray(input, mid+1, high)
        cl, ch, cs = find_crossing_subarray(input, low, mid, high)

        #If left sum is >= right sum and cross sum, return the left subarray vals
        if(ls >= rs and ls >= cs):
            return(ll, lh, ls)
        #If right sum > left sum and cross sum, return the right subarray vals
        elif (rs > ls and rs > cs):
            return(rl, rh, rs)
        #Else return the crossing subarray vals
        else:
            return(cl, ch, cs)

def find_crossing_subarray(input, low, mid, high):
    #Find the max subarray starting at mid down to low
    leftsum = -math.inf
    sum = 0
    maxleft = mid
    for i in range(mid, low-2, -1):
        sum += input[i]
        if (sum > leftsum): 
            leftsum = sum
            maxleft = i

    #Find the max subarray starting at mid up to high
    rightsum = -math.inf
    sum = 0
    maxright = mid+1
    for i in range(mid+1, high):
        sum += input[i]
        if (sum > rightsum):
            rightsum = sum
            maxright = i
     
    return(maxleft, maxright, leftsum+rightsum)

stockdataframe = pd.read_excel("stockdata.xlsx", header=None)
value = stockdataframe.loc[:,1]
dates = stockdataframe.loc[:,0]


valChange = [0] * 40
for i in range(1, 39):
    valChange[i] = round(value.loc[i] - value.loc[i-1],2)

#valChange array now holds the differences between that day and the previous day's stock market value
buyIdx, sellIdx, gain = find_max_subarray(valChange, 0, 39)

#get ideal buy and sell dates from the dates column from the buy and sell indices
buyDay = dates.loc[buyIdx-1]
sellDay = dates.loc[sellIdx]
print(buyDay, sellDay)

#buyDayVal is value at day of buy, used to calculate # of stocks you buy, sellDayVal is value on day to sell
buyDayVal = value.loc[buyIdx-1]
numStocks = 1000/buyDayVal
sellDayVal = value.loc[sellIdx]

#ending value is the number of stocks * the value on day of sell, profit is difference between initial cost and endVal
endVal = numStocks * sellDayVal
profit = endVal - 1000.00
print(round(profit,2))








    
    

    
```
